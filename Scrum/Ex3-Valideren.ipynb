```python
import math

X1=P1.Cgraden
X2=P2.Cgraden
X4=P4.Cgraden
X5=P5.Cgraden


gemiddeld_X1= sum(P1.Cgraden)/len(P1.Cgraden)
gemiddeld_X2= sum(P2.Cgraden)/len(P2.Cgraden)
gemiddeld_X4= sum(P4.Cgraden)/len(P4.Cgraden)
gemiddeld_X5= sum(P5.Cgraden)/len(P5.Cgraden)

Y1=P1.Kgraden
gemiddeld_Y1= sum(P1.Kgraden)/len(P1.Kgraden)

Y2=P2.Kgraden
gemiddeld_Y2= sum(P2.Kgraden)/len(P2.Kgraden)

Y4=P4.Kgraden
gemiddeld_Y4= sum(P4.Kgraden)/len(P4.Kgraden)

Y5=P5.Kgraden
gemiddeld_Y5= sum(P5.Kgraden)/len(P5.Kgraden)


X1= (X1-gemiddeld_X1)
Y1= (Y1-gemiddeld_Y1)

X2= (X2-gemiddeld_X2)
Y2= (Y2-gemiddeld_Y2)

X4= (X4-gemiddeld_X4)
Y4= (Y4-gemiddeld_Y4)

X5= (X5-gemiddeld_X5)
Y5= (Y5-gemiddeld_Y5)


corelatieCoefficient1= sum(X1*Y1)/ math.sqrt(sum(X1*X1)*sum(Y1*Y1))
corelatieCoefficient2= sum(X2*Y2)/ math.sqrt(sum(X2*X2)*sum(Y2*Y2))
corelatieCoefficient4= sum(X4*Y4)/ math.sqrt(sum(X4*X4)*sum(Y4*Y4))
corelatieCoefficient5= sum(X5*Y5)/ math.sqrt(sum(X5*X5)*sum(Y5*Y5))

print('CorelatieCoeeficient P1E3: ' +str(corelatieCoefficient1))
print('CorelatieCoeeficient P2E3: ' +str(corelatieCoefficient2))
print('CorelatieCoeeficient P4E3: ' +str(corelatieCoefficient4))
print('CorelatieCoeeficient P5E3: ' +str(corelatieCoefficient5))

```
