# Correlatie Cooficient

```python
import math

X1=P1.Dgraden
X2=P2.Dgraden
X4=P4.Dgraden
X5=P5.Dgraden

Y1=P1.RechtsGrd
Y2=P2.RechtsGrd
Y4=P4.RechtsGrd
Y5=P5.RechtsGrd

gemiddeld_X1= sum(P1.Dgraden)/len(P1.Dgraden)
gemiddeld_X2= sum(P2.Dgraden)/len(P2.Dgraden)
gemiddeld_X4= sum(P4.Dgraden)/len(P4.Dgraden)
gemiddeld_X5= sum(P5.Dgraden)/len(P5.Dgraden)


gemiddeld_Y1= sum(P1.RechtsGrd)/len(P1.RechtsGrd)
gemiddeld_Y2= sum(P2.RechtsGrd)/len(P2.RechtsGrd)
gemiddeld_Y4= sum(P4.RechtsGrd)/len(P4.RechtsGrd)
gemiddeld_Y5= sum(P5.RechtsGrd)/len(P5.RechtsGrd)


X1= (X1-gemiddeld_X1)
Y1= (Y1-gemiddeld_Y1)

X2= (X2-gemiddeld_X2)
Y2= (Y2-gemiddeld_Y2)

X4= (X4-gemiddeld_X4)
Y4= (Y4-gemiddeld_Y4)

X5= (X5-gemiddeld_X5)
Y5= (Y5-gemiddeld_Y5)

corelatieCoefficient1= sum(X1 * Y1)/ math.sqrt(sum(X1 * X1)*sum(Y1 * Y1))
corelatieCoefficient2= sum(X2 * Y2)/ math.sqrt(sum(X2 * X2)*sum(Y2 * Y2))
corelatieCoefficient4= sum(X4 * Y4)/ math.sqrt(sum(X4 * X4)*sum(Y4 * Y4))
corelatieCoefficient5= sum(X5 * Y5)/ math.sqrt(sum(X5 * X5)*sum(Y5 * Y5))

print('CorelatieCoeeficient P1E1: ' +str(corelatieCoefficient1))
print('CorelatieCoeeficient P2E1: ' +str(corelatieCoefficient2))
print('CorelatieCoeeficient P4E1: ' +str(corelatieCoefficient4))
print('CorelatieCoeeficient P5E1: ' +str(corelatieCoefficient5))
```
